```python
# hora y fecha actual, para calcular los días desde el registro y la última actividad
today = pd.Timestamp(datetime.today())
# Convertir las columnas de fecha a tipo datetime y calcular los días desde el registro y la última actividad
df['signup_date'] = pd.to_datetime(df['signup_date'])
df['last_activity_date'] = pd.to_datetime(df['last_activity_date'])
# Calcular los días desde el registro y la última actividad
df['days_since_signup'] = (today - df['signup_date']).dt.days
df['days_since_last_activity'] = (today - df['last_activity_date']).dt.days
# eliminar las columnas de fecha originales
df.drop(['signup_date', 'last_activity_date'], axis=1, inplace=True)
```

```python
df.head()
```

```python
bool_cols = ['used_discount', 'auto_renew', 'has_multiple_profiles']
df[bool_cols] = df[bool_cols].astype(int)
```

```python
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

```python
df = pd.get_dummies(df, columns=[
    'subscription_type', 'payment_method', 'country', 'device_type'
], drop_first=True)
```
